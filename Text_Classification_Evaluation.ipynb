```python
# Import Libraries
import pandas as pd
import numpy as np
import nltk
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
# Load the course data files for English into a dataframe using Pandas
dataframe_english = pd.read_csv(r"C:\Users\muthu\Desktop\UNCC\Semester-3\NLP\Assignments\Data_Input\CONcreTEXT_trial_EN.tsv", sep = "\t")
dataframe_english
```

```python
# Load the course data files for Italian into a dataframe using Pandas
dataframe_italian = pd.read_csv(r"C:\Users\muthu\Desktop\UNCC\Semester-3\NLP\Assignments\Data_Input\CONcreTEXT_trial_IT.tsv", sep = "\t")
dataframe_italian
```

```python
# Combining both the dataframes using concat function in pandas
combined_dataframe = pd.concat([dataframe_english, dataframe_italian])
combined_dataframe
```

## Question-1 Answer - Assigning Class Labels for the combined Dataframe

```python
# Question-1
condition = [combined_dataframe['MEAN'] <= 4, combined_dataframe['MEAN'] > 4]
class_labels = ["LOW", 'HIGH']
    
combined_dataframe['CONCRETE'] = np.select(condition, class_labels)
combined_dataframe
```

<a id="Question 2 Split the Dataset">
</a>

## Question-2 Answer - Split the Dataset

```python
combined_dataframe['LEN_OF_CHAR_TARGET'] = combined_dataframe['TARGET'].str.len()
combined_dataframe
```

```python
combined_dataframe['TARGET_str_len_concrete'] = np.where(combined_dataframe['TARGET'].str.len() >= 5, 'HIGH', 'LOW')
combined_dataframe
```

```python
# Spliting the training and test data set in 80% : 20% using train_test_split

combined_df_train, combined_df_test = train_test_split(combined_dataframe, test_size=0.2, shuffle=True, random_state=42)

print("\t English Language Train data: \n\n", combined_df_train.head(5), "\n\n Length of the Train data is: ", len(combined_df_train))
print('\n')
print("***********************************************************")
print('\n')
print("\t English Language Test data: \n\n", combined_df_test.head(5), "\n\n Length of the Train data is: ", len(combined_df_test))
```

## Majority Class Baseline

```python
from sklearn.dummy import DummyClassifier
X = combined_df_train['TEXT']
y = combined_df_train['CONCRETE']
dummy_clf = DummyClassifier(strategy="most_frequent")
```

```python
dummy_clf.fit(X, y)
```

```python
predict_value = dummy_clf.predict(combined_df_test['CONCRETE'])
predict_value
```

```python
target_names = combined_df_train['CONCRETE'].unique()
target_names
```

## Question-3 Answer - Majority Class Baseline

```python
print(metrics.classification_report(combined_df_test['CONCRETE'], predict_value,
    target_names=target_names))

metrics.confusion_matrix(combined_df_test['CONCRETE'], predict_value)
```

## Predict whether the text is HIGH concrete or LOW concrete

```python
# CountVectorizer() is used for Text preprocessing, tokenizing and filtering of stopwords in the given text.
count_vect = CountVectorizer()
combined_train_data_count_vect = count_vect.fit_transform(combined_df_train['TEXT'])
combined_train_data_count_vect.shape
```

```python
combined_df_train['TEXT']
```

```python
# TfidfTransformer() is used for converting the words occurrences to frequencies
tfidf_transformer = TfidfTransformer()
combined_train_data_count_vect_tfidf = tfidf_transformer.fit_transform(combined_train_data_count_vect)
combined_train_data_count_vect_tfidf.shape
```

```python
naive_bayes_model = MultinomialNB().fit(combined_train_data_count_vect_tfidf, combined_df_train['CONCRETE'])
naive_bayes_model
```

<a id="Question 5 Answer_predict_HIGH_LOW">
</a>

## Question-5 Answer - Predict whether the text is HIGH concrete or LOW concrete

```python
category = ['HIGH', 'LOW']
docs_new = combined_df_test['TEXT']
X_new_counts = count_vect.transform(docs_new)
X_new_tfidf = tfidf_transformer.transform(X_new_counts)

predicted = naive_bayes_model.predict(X_new_tfidf)
predicted

for doc, category in zip(docs_new, predicted):
    #print(predicted)
    print('%r => %s' % (doc, category))
```

```python
category = ['HIGH', 'LOW']
print(metrics.classification_report(combined_df_test['CONCRETE'], predicted,
    target_names=category))

metrics.confusion_matrix(combined_df_test['CONCRETE'], predicted)
```
