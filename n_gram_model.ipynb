## Question-2 N-gram Function

```python
# General Ngram Function
def ngram_function(ngram, input_string):
    ngra_eng = [input_string[i:i+ngram] for i in range(len(input_string)-ngram+1)]
    return ngra_eng
```

```python
# Unigram Output for English
unigram_english = ngram_function(1, english_series_to_string)
unigram_english
```

## Question-2 N-gram model Implementation

```python
# Function for Unigram English Test set

# Input is given as list
def Unigram_English(input_test_set):
    count_eng = 0                                # Accuracy for English words
    count_ita = 0                                # Accuracy for Italian words
#     total_english_prob = 0
#     total_italian_prob = 0
    # Iterating through each item in the English test set
    for every_item in input_test_set:                  # 'every_item' have individual words from the list-Bring
        total_english_prob = 0
        total_italian_prob = 0
        char_ngram = ngram_function(1, every_item)     # 'char_ngram' have individual characters in the word obtained from prev step-['B', 'r', 'i', 'n', 'g'] 
        for char_index in range(0, len(char_ngram)):   # 'char_index' have the index of the individual charcters of prev step
            # Iterating over English Unigram Frequent items
            for ele in english_unigram_fdist:          
                if ele == (char_ngram[char_index]):
                    if ele in english_final_uni_dict.keys():
                        total_english_prob = total_english_prob + english_final_uni_dict[ele]
                    else:
                        break
            for ele in italian_unigram_fdist:
                if ele == (char_ngram[char_index]):
                    if ele in italian_final_uni_dict.keys():
                        total_italian_prob = total_italian_prob + italian_final_uni_dict[ele]
                    else:
                        break
        print("Total English Probability for the word is: ", total_english_prob)
        print("Total Italian Probability for the word is: ", total_italian_prob)
        if total_english_prob > total_italian_prob:
            print (every_item.upper()," is an English word!")
            print("\n")
        else: 
            print (every_item.upper()," is an Italian word!")   
            print("\n")
#         count_eng = 0
#         count_ita = 0
        if total_english_prob > total_italian_prob:
            count_eng = count_eng + 1
        else: 
            count_ita = count_ita + 1
#         return count_eng, count_ita
        print("Total count of Predicted English words: ", count_eng)
        print("Total count of Predicted Italian words: ", count_ita)
        english_accuracy = (count_eng/len(english_test_set_final_list))*100
        italian_accuracy = (count_ita/len(italian_test_set_final_list))*100
    print("Accuracy for English language is: ", english_accuracy)    
    print("Accuracy for Italian language is: ", italian_accuracy)   
```
