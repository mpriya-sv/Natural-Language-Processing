```python
import pandas as pd
import numpy as np
import nltk
import random
from nltk.lm.models import Laplace
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import KneserNeyInterpolated
import string
from nltk.util import pad_sequence
from nltk.util import bigrams
from nltk.util import ngrams
from nltk.util import everygrams
```

```python
# Question-1
# Load the course data files for English into a dataframe using Pandas
dataframe_english = pd.read_csv(r"C:\Users\muthu\Desktop\UNCC\Semester-3\NLP\Assignments\Data_Input\CONcreTEXT_trial_EN.tsv", sep = "\t")
dataframe_english
```

```python
eng_corpus = list(dataframe_english['TEXT'])
eng_corpus
```

```python
# Joining the Series in English as String
english_series_to_string = " ".join(eng_corpus)
english_series_to_string
```

```python
eng_split_token = english_series_to_string.split()
eng_split_token
```

```python
random.shuffle(eng_split_token)
print(eng_split_token[:5])
```

```python
len(eng_split_token)
```

```python
en_train = [w.lower() for w in eng_split_token][:int(len(eng_split_token)*.8)]
en_test = [w.lower() for w in eng_split_token][int(len(eng_split_token)*.8):]
```

```python
en_train_string = " ".join(en_train)
en_train_string
```

```python
en_train
```

```python
print("Length of English Train is: ", len(en_train))
print("Length of English Test is: ", len(en_test))
```

```python
from nltk.util import pad_sequence
padded_sent_bi = list(pad_sequence(en_train_string,
                  pad_left=True, left_pad_symbol="<s>",
                  pad_right=True, right_pad_symbol="</s>",
                  n=2)) # The n order of n-grams, if it's 2-grams, you pad once, 3-grams pad twice, etc. 
padded_sent_bi
```

```python
ngram_pad_bi = list(everygrams(padded_sent_bi, max_len=2))
ngram_pad_bi
```

```python
padded_sent_list_bi = [ngram_pad_bi]
padded_sent_list_bi
```

```python
vocab = list(en_train_string)
vocab
```

<a id="Question 1 Answer English Language - Bigram - Laplace Smoothing">
</a>

# English Language - Bigram - Laplace Smoothing

```python
lap_model_bi = Laplace(2)
lap_model_bi.fit(padded_sent_list_bi, vocab)
```

```python
len(lap_model_bi.vocab)
#len(lm.vocab)
```

```python
lap_model_bi.counts['b']
```

```python
lap_model_bi.counts['s']
```

```python
lap_model_bi.counts[['b']]['a']
```

```python
lap_model_bi.counts[['s']]['u']
```

```python
lap_model_bi.score('bring')
```

```python
lap_model_bi.score('academic')
```

```python
lap_model_bi.generate(20, ['b'])
```

```python
lap_model_bi.generate(20)
```

```python
print(''.join(lap_model_bi.generate(1000, random_seed = 7)))
```
